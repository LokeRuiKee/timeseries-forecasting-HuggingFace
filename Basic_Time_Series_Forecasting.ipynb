```python
import numpy as np

# Define a function to calculate moving average
def moving_average(series, window_size):
    """
    Calculate the moving average of a time series.

    Args:
        series (list): The time series data.
        window_size (int): The size of the moving window.

    Returns:
        np.array: The moving averages.
    """
    return np.convolve(series, np.ones(window_size) / window_size, mode='valid')

# Define window size for moving average
window_size = 5

# Calculate moving average for each time series
ma_T1 = moving_average(T1, window_size)
ma_T2 = moving_average(T2, window_size)
ma_T3 = moving_average(T3, window_size)

# Plot original data and moving averages
plt.figure(figsize=(10, 6))
plt.plot(range(len(T1)), T1, label='T1 (Original)')
plt.plot(range(window_size - 1, len(T1)), ma_T1, label=f'T1 (Moving Average, window={window_size})')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series T1 with Moving Average Forecast')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(range(len(T2)), T2, label='T2 (Original)')
plt.plot(range(window_size - 1, len(T2)), ma_T2, label=f'T2 (Moving Average, window={window_size})')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series T2 with Moving Average Forecast')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(range(len(T3)), T3, label='T3 (Original)')
plt.plot(range(window_size - 1, len(T3)), ma_T3, label=f'T3 (Moving Average, window={window_size})')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Time Series T3 with Moving Average Forecast')
plt.legend()
plt.show()
```

```python
# Pipeline automation

import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from datetime import datetime, timedelta

# Step 1: Data Collection (Replace this with your actual data collection mechanism)
def collect_data():
    # Simulated load data for demonstration purposes
    start_date = datetime(2024, 4, 1)
    end_date = datetime(2024, 4, 7)
    timestamps = pd.date_range(start=start_date, end=end_date, freq='15T')
    load_values = [100 + 10 * i + 5 * (i % 24) for i in range(len(timestamps))]
    load_data = pd.DataFrame({'timestamp': timestamps, 'load': load_values})
    return load_data

# Step 2: Data Preprocessing
def preprocess_data(load_data):
    # No preprocessing required in this example
    return load_data

# Step 3: Model Training
def train_model(train_data):
    # Simple Random Forest Regressor for demonstration purposes
    X_train = pd.to_numeric(train_data['timestamp']).values.reshape(-1, 1)
    y_train = train_data['load'].values
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    return model

# Step 4: Forecasting
def forecast_load(model, last_timestamp):
    # Generate timestamps for the next 24 hours (96 periods of 15 minutes)
    forecast_timestamps = pd.date_range(start=last_timestamp + timedelta(minutes=15), periods=96, freq='15T')
    # Convert timestamps to numeric format for prediction
    X_forecast = pd.to_numeric(forecast_timestamps).values.reshape(-1, 1)
    # Make load forecasts using the trained model
    load_forecast = model.predict(X_forecast)
    # Create a DataFrame to store the forecasts
    forecast_data = pd.DataFrame({'timestamp': forecast_timestamps, 'load_forecast': load_forecast})
    return forecast_data

# Step 5: Deployment (Assuming this function is called periodically, e.g., every 15 minutes)
def main():
    # Step 1: Data Collection
    load_data = collect_data()

    # Step 2: Data Preprocessing
    load_data = preprocess_data(load_data)

    # Step 3: Model Training
    model = train_model(load_data)

    # Step 4: Forecasting
    last_timestamp = load_data['timestamp'].max()
    forecast_data = forecast_load(model, last_timestamp)

    # Print or save the forecast data
    print(forecast_data.head())

if __name__ == "__main__":
    main()
```
